```python
import matplotlib as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("diabetes.csv")

diabetes_column = data[[
    "Pregnancies",
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI",
    "DiabetesPedigreeFunction",
    "Age",
    "Outcome"
]]
diabetes_color = {1 : 'red', 
                0: 'blue'}
sns.pairplot(diabetes_column, hue='Outcome', palette=diabetes_color)

plt.show()
```
